```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
import mglearn
```

```python
iris_dataset = datasets.load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataset.data, iris_dataset.target, test_size=0.8, random_state=0)
```

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, c = y_train, figsize=(15,15),marker='o',hist_kwds={'bins':20},s=60,alpha=0.8,cmap=mglearn.cm3)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
KNeighborsClassifier(n_neighbors=1)
```

```python
prediction = knn.predict(X_test)
```

```python
print("Predicted\tActual")
for Predicted, Actual in zip(prediction, y_test):
    print(
        f"{iris_dataset["target_names"][Predicted]}\t{iris_dataset["target_names"][Actual]}"
    )
```
